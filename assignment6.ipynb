```python
import keras
import pandas
from keras.models import Sequential
from sklearn.datasets import load_breast_cancer
from keras.layers.core import Dense, Activation

# load dataset
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np

# Load dataset
df = load_breast_cancer()

# Separate the target variable from the input features
X = df.data
Y = df.target

# Normalize the input features
sc = StandardScaler()
X = sc.fit_transform(X)

# Split the dataset into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=87)


np.random.seed(155)
my_first_nn = Sequential() # create model
my_first_nn.add(Dense(20, input_dim=30, activation='relu')) # hidden layer
my_first_nn.add(Dense(1, activation='sigmoid')) # output layer
my_first_nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['acc'])

# Train the model
my_first_nn_fitted = my_first_nn.fit(X_train, Y_train, epochs=100, initial_epoch=0)

# Evaluate the model
print(my_first_nn.summary())
print(my_first_nn.evaluate(X_test, Y_test))

loss, accuracy = my_first_nn.evaluate(X_test, Y_test)
print("Accuracy:", accuracy)
```

Here after normalizing accuracy increase to 0.96 from 0.84

```python
from keras import Sequential
from keras.datasets import mnist
import numpy as np
from keras.layers import Dense
from tensorflow.keras.utils import to_categorical

(train_images,train_labels),(test_images, test_labels) = mnist.load_data()

print(train_images.shape[1:])

#process the data
#convert each image of shape 28*28 to 784 dimensional which will be fed to the network as a single feature
dimData = np.prod(train_images.shape[1:])
print(dimData)
train_data = train_images.reshape(train_images.shape[0],dimData)/ 255
test_data = test_images.reshape(test_images.shape[0],dimData)/ 255

#1 plot one of the images in the test data
import matplotlib.pyplot as plt
plt.imshow(test_images[0])
plt.show()

#2. inferencing on a single image
test_img = test_data[0].reshape(1,-1)
model = Sequential()
pred = model.predict(test_img)
print("Prediction:", np.argmax(pred))

#convert data to float and scale values between 0 and 1
train_data = train_data.astype('float32')
test_data = test_data.astype('float32')

#3. change the labels from integer to one-hot encoding. to_categorical is doing the same thing as LabelEncoder()
train_labels_one_hot = to_categorical(train_labels)
test_labels_one_hot = to_categorical(test_labels)

#4. Run the same code without scaling the images
train_data_raw = train_images.reshape(train_images.shape[0],dimData)
test_data_raw = test_images.reshape(test_images.shape[0],dimData)

#creating network
#3. change the activation function to tanh

model.add(Dense(512, activation='tanh', input_shape=(dimData,)))
#3. change the number of hidden layers to 3
model.add(Dense(256, activation='tanh'))
model.add(Dense(128, activation='tanh'))
model.add(Dense(10, activation='softmax'))

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

#4. Train the model without scaling the images
history_raw = model.fit(train_data_raw, train_labels_one_hot, batch_size=256, epochs=10, verbose=1,
                        validation_data=(test_data_raw, test_labels_one_hot))

#training the model
history = model.fit(train_data, train_labels_one_hot, batch_size=256, epochs=10, verbose=1,
                   validation_data=(test_data, test_labels_one_hot))

#1.2 plot the loss and accuracy for both training data and validation data
import matplotlib.pyplot as plt
print("With out scalling :")
plt.plot(history_raw.history['accuracy'])
plt.plot(history_raw.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()

plt.plot(history_raw.history['loss'])
plt.plot(history_raw.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()

print("With scalling :")

plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()

```

For the second one i have made all the points in one  code for the first i plotted one of the images from the MNIST dataset to visualize what the data looks like. then code runs the trained neural network on a single image from the test dataset and predicts the label of the image . after code modifies the neural network architecture by changing the activation function of the hidden layers to tanh and adding one more hidden layer.This code trains and evaluates the neural network without scaling the image data, to compare the performance of the model with and without data normalization.This code plots the training and validation loss and accuracy values over the epochs to visualize how the model is improving during training.
